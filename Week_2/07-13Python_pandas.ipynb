## DataFrame 생성
### DataFrame : Series들이 모인 2차원 데이터
* Dictionary들의 list로부터 생성 가능



```python
import numpy as np
import pandas as pd


rawData = {"3일" : [3, 5, 10, 2], "4일" : [10, 9, 0, 1], "5일" : [5, 7,6, 3]}
data = pd.DataFrame(rawData)
display(data)
```

```python
# index 부여하기

data.index = ["치킨", "비프", "삼겹살", "함박"]
display(data)
```

```python
# 열 선택하기
display(data[["3일"]])

display(data.get(["3일"]))
```

```python
## 열 선택하기
### 열들이 붙어있지 않아도 선택 가능

display(data[["3일", "5일"]])
```

```python
# 행 선택하기
## Array 선택하듯이 선택 가능

print(data[0:2])

print(data[0::2])

## index를 사용해도 됨
### 이 경우에는 끝 인덱스를 포함한다.

print(data["치킨":"삼겹살":2])
```

```python
# 행/열 선택하기

## iloc : 숫자로 선택하기

print(data.iloc[::2, 1])
print("")
print(data.iloc[3, 2])
print("")
## loc : column명 & index로 선택하기

print(data.loc[:"삼겹살", "4일"])
```

```python
# 데이터 필터링
print(data[data["5일"] > 4])

# 데이터 정렬
print(data.sort_values("5일"))
print(data.sort_values("5일", ascending=False))

# 데이터 기초통계량
print(data.describe())
```

```python
# 엑셀에 쓰기

data.to_excel("dosirak.xlsx")

# 엑셀에서 읽기
data.to_excel("dosirak.xlsx")
ex = pd.read_excel("dosirak.xlsx")
print(type(ex))
print(ex)
```

## 실습 - 꺾은선 그래프 
* 𝑦(𝑥) = 𝑥 𝑥 − 2  
* 𝑧(𝑥) = −𝑥 𝑥 − 2  
* x는 -10부터 10까지 정수  
* 위의 두 식을 변수들로 표현  
* 같은 그래프에 그리기

```python

x = np.arange(-10, 11)
y = x*(x-2)
z = -x*(x-2)

plt.plot(x, y)
plt.plot(x, z)
plt.legend(['y = x*(x-2)',' z = -x*(x-2)'])
plt.show()
```

### 점/선 모양 바꾸기

```python
plt.plot(x, y, '-b')
plt.plot(x, z, 'r:x')
plt.legend(['y = x*(x-2)',' z = -x*(x-2)'])
plt.show()
```

### 데이터에 이름 붙이기


```python
plt.title('test graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x, y, "-b", label="y values")
plt.plot(x, z, "r:x", label="z values")
plt.legend()
plt.show()
```

### 범위 지정 및 눈금 활성화

```python
x = np.arange(-10 ,11, 0.1)
y = x*(x-2)
z = -x*(x-2)

plt.title('test graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(-2, 4)
plt.ylim(-5, 5)
plt.plot(x, y, "-b", label="y values")
plt.plot(x, z, "r:x", label="z values")
plt.grid(True)
plt.legend()
plt.show()
```

### 막대 그래프 그리기


```python
x = np.arange(-10 ,11)
y = x*(x-2)
z = -x*(x-2)

plt.bar(x, y, color='red')
plt.plot(x, y)
plt.show()
```

### 원 그래프 그리기


```python
plt.rcParams['figure.figsize'] = [12, 8]


group_names = ['Group_A', 'Group_B', 'Group_C']

group_sizes = [95, 54, 25]

group_colors = ['yellowgreen', 'lightskyblue', 'lightcoral']


plt.pie(group_sizes, 
        labels=group_names, 
        colors=group_colors, 
        autopct='%1.2f%%',
        shadow=True, 
        startangle=90,
        textprops={'fontsize': 14}) 

plt.axis('equal') 

plt.title('Pie Chart of Market Share', fontsize=20)

plt.show()


```

### 히스토그램 그리기


```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X, Y = cancer.data[:, 0], cancer.target

X0 = X[Y == 0] # Y가 0인 값을 분리
X1 = X[Y == 1] # Y가 1인 값을 분리

plt.hist(X0, bins=20, color='red', alpha=.8)
plt.hist(X1, bins=20, color='green', alpha=.8)
plt.show()
```
