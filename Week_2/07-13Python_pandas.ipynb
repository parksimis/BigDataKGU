## DataFrame ÏÉùÏÑ±
### DataFrame : SeriesÎì§Ïù¥ Î™®Ïù∏ 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞
* DictionaryÎì§Ïùò listÎ°úÎ∂ÄÌÑ∞ ÏÉùÏÑ± Í∞ÄÎä•



```python
import numpy as np
import pandas as pd


rawData = {"3Ïùº" : [3, 5, 10, 2], "4Ïùº" : [10, 9, 0, 1], "5Ïùº" : [5, 7,6, 3]}
data = pd.DataFrame(rawData)
display(data)
```

```python
# index Î∂ÄÏó¨ÌïòÍ∏∞

data.index = ["ÏπòÌÇ®", "ÎπÑÌîÑ", "ÏÇºÍ≤πÏÇ¥", "Ìï®Î∞ï"]
display(data)
```

```python
# Ïó¥ ÏÑ†ÌÉùÌïòÍ∏∞
display(data[["3Ïùº"]])

display(data.get(["3Ïùº"]))
```

```python
## Ïó¥ ÏÑ†ÌÉùÌïòÍ∏∞
### Ïó¥Îì§Ïù¥ Î∂ôÏñ¥ÏûàÏßÄ ÏïäÏïÑÎèÑ ÏÑ†ÌÉù Í∞ÄÎä•

display(data[["3Ïùº", "5Ïùº"]])
```

```python
# Ìñâ ÏÑ†ÌÉùÌïòÍ∏∞
## Array ÏÑ†ÌÉùÌïòÎìØÏù¥ ÏÑ†ÌÉù Í∞ÄÎä•

print(data[0:2])

print(data[0::2])

## indexÎ•º ÏÇ¨Ïö©Ìï¥ÎèÑ Îê®
### Ïù¥ Í≤ΩÏö∞ÏóêÎäî ÎÅù Ïù∏Îç±Ïä§Î•º Ìè¨Ìï®ÌïúÎã§.

print(data["ÏπòÌÇ®":"ÏÇºÍ≤πÏÇ¥":2])
```

```python
# Ìñâ/Ïó¥ ÏÑ†ÌÉùÌïòÍ∏∞

## iloc : Ïà´ÏûêÎ°ú ÏÑ†ÌÉùÌïòÍ∏∞

print(data.iloc[::2, 1])
print("")
print(data.iloc[3, 2])
print("")
## loc : columnÎ™Ö & indexÎ°ú ÏÑ†ÌÉùÌïòÍ∏∞

print(data.loc[:"ÏÇºÍ≤πÏÇ¥", "4Ïùº"])
```

```python
# Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
print(data[data["5Ïùº"] > 4])

# Îç∞Ïù¥ÌÑ∞ Ï†ïÎ†¨
print(data.sort_values("5Ïùº"))
print(data.sort_values("5Ïùº", ascending=False))

# Îç∞Ïù¥ÌÑ∞ Í∏∞Ï¥àÌÜµÍ≥ÑÎüâ
print(data.describe())
```

## Ïã§Ïäµ - Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ 
* ùë¶(ùë•) = ùë• ùë• ‚àí 2  
* ùëß(ùë•) = ‚àíùë• ùë• ‚àí 2  
* xÎäî -10Î∂ÄÌÑ∞ 10ÍπåÏßÄ Ï†ïÏàò  
* ÏúÑÏùò Îëê ÏãùÏùÑ Î≥ÄÏàòÎì§Î°ú ÌëúÌòÑ  
* Í∞ôÏùÄ Í∑∏ÎûòÌîÑÏóê Í∑∏Î¶¨Í∏∞

```python

x = np.arange(-10, 11)
y = x*(x-2)
z = -x*(x-2)

plt.plot(x, y)
plt.plot(x, z)
plt.legend(['y = x*(x-2)',' z = -x*(x-2)'])
plt.show()
```

### Ï†ê/ÏÑ† Î™®Ïñë Î∞îÍæ∏Í∏∞

```python
plt.plot(x, y, '-b')
plt.plot(x, z, 'r:x')
plt.legend(['y = x*(x-2)',' z = -x*(x-2)'])
plt.show()
```

### Îç∞Ïù¥ÌÑ∞Ïóê Ïù¥Î¶Ñ Î∂ôÏù¥Í∏∞


```python
plt.title('test graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x, y, "-b", label="y values")
plt.plot(x, z, "r:x", label="z values")
plt.legend()
plt.show()
```

### Î≤îÏúÑ ÏßÄÏ†ï Î∞è ÎààÍ∏à ÌôúÏÑ±Ìôî

```python
x = np.arange(-10 ,11, 0.1)
y = x*(x-2)
z = -x*(x-2)

plt.title('test graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(-2, 4)
plt.ylim(-5, 5)
plt.plot(x, y, "-b", label="y values")
plt.plot(x, z, "r:x", label="z values")
plt.grid(True)
plt.legend()
plt.show()
```

### Ïõê Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞


```python
plt.rcParams['figure.figsize'] = [12, 8]


group_names = ['Group_A', 'Group_B', 'Group_C']

group_sizes = [95, 54, 25]

group_colors = ['yellowgreen', 'lightskyblue', 'lightcoral']


plt.pie(group_sizes, 
        labels=group_names, 
        colors=group_colors, 
        autopct='%1.2f%%',
        shadow=True, 
        startangle=90,
        textprops={'fontsize': 14}) 

plt.axis('equal') 

plt.title('Pie Chart of Market Share', fontsize=20)

plt.show()


```

### ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏Î¶¨Í∏∞


```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X, Y = cancer.data[:, 0], cancer.target

X0 = X[Y == 0] # YÍ∞Ä 0Ïù∏ Í∞íÏùÑ Î∂ÑÎ¶¨
X1 = X[Y == 1] # YÍ∞Ä 1Ïù∏ Í∞íÏùÑ Î∂ÑÎ¶¨

plt.hist(X0, bins=20, color='red', alpha=.8)
plt.hist(X1, bins=20, color='green', alpha=.8)
plt.show()
```
